```python
import pickle
import pandas as pd
```

# type 1 : all combined and mixed without checks
here we have done thresholding of 0.8 for both the cols, and then we take a sum of the cosine similarities, and pick the top 2000

```python
with open("/content/drive/MyDrive/hcml/with_all_RM_data/Iteration_1/prompts_sims_iter1.pkl", "rb") as f:
  data = pickle.load(f)
```

```python
data[0]
```

```python
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
df2 = df.drop('pairs', axis=1)
```

```python
df2 = df2.drop('masks', axis=1)
```

```python
df2.columns
```

```python
df2 = df2[(df2['similarity_with_chosen'] > 0.8) & (df2['similarity_with_rejected'] > 0.8)]
```

```python
df2.head()
```

```python
df2['sum'] = df2['similarity_with_chosen'] + df2['similarity_with_rejected']
```

```python
df2
```

```python
df2 = df2.sort_values(by=['sum'])
```

```python
df2
```

```python
subset = df2.iloc[-2200:]
subset
```

```python
df_part_1 = df.iloc[subset.index]
```

```python
df_part_1.iloc[:10]
```

```python
def clean_up(row):
   row['similarity_with_chosen'] = row['similarity_with_chosen'][0].item()
   row['similarity_with_rejected'] = row['similarity_with_rejected'][0].item()
   return row
```

```python
df_part_1 = df_part_1.apply(clean_up, axis=1)
```

```python
df_part_1.iloc[:5]
```

```python
len(df_part_1)
```

```python
# remove random 200 to make test set
train=df_part_1.sample(2000)
test=df_part_1.drop(train.index)
```

```python
len(test)
```

```python
part_1_train = train.to_dict('records')
part_1_test = test.to_dict('records')
```

```python
pickle.dump(part_1_train, file=open('/content/drive/MyDrive/hcml/with_all_RM_data/Iteration_1/frft_part1_train.pkl', 'wb'))
pickle.dump(part_1_test, file=open('/content/drive/MyDrive/hcml/with_all_RM_data/Iteration_1/frft_part1_test.pkl', 'wb'))
```
